```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import glob
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.model_selection import train_test_split
```

```python
# Chemins des fichiers
BASE_PATH = r"C:\Users\TOURE\Desktop\Projet_Rakuten"
IMAGE_FOLDER = os.path.join(BASE_PATH, "image_train")
TRAIN_CSV = os.path.join(BASE_PATH, "X_train.csv")
LABEL_CSV = os.path.join(BASE_PATH, "Y_train.csv")
```

```python
# Nombre d'images à charger
sample_size = 1000
```

```python
# Chargement des données
train_df = pd.read_csv(TRAIN_CSV)
label_df = pd.read_csv(LABEL_CSV)
data = pd.concat([train_df, label_df], axis=1)
```

```python
# 1. Réduction de la taille des images
def load_and_resize_image(image_id, target_size=(64, 64)):
    image_path = os.path.join(IMAGE_FOLDER, f"image_{image_id}_product_*.jpg")
    image_files = glob.glob(image_path)
    if not image_files:
        print(f"Aucune image trouvée pour l'ID: {image_id}")
        return np.zeros(target_size + (3,))
    image_file = image_files[0]
    img = Image.open(image_file).convert('RGB')
    img = img.resize(target_size)
    return np.array(img)

# Chargement d'un sous-ensemble d'images
X = []
y = []
for index, row in data.iterrows():
    if len(X) >= sample_size:
        break
    img = load_and_resize_image(row['imageid'])
    X.append(img)
    y.append(row['prdtypecode'])
    if len(X) % 10 == 0:
        print(f"Chargé {len(X)} images")

X = np.array(X)
y = np.array(y)

print(f"Chargement terminé. {len(X)} images chargées.")
```

```python
# 3. Équilibrage des classes avec data augmentation
datagen = ImageDataGenerator(
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    horizontal_flip=True,
    zoom_range=0.2
)
```

```python
# Affichage de quelques images traitées
plt.figure(figsize=(10, 5))
for i in range(min(5, len(X))):
    plt.subplot(1, 5, i+1)
    plt.imshow(X[i])
    plt.title(f"Class: {y[i]}")
    plt.axis('off')
plt.show()
```

```python
# 2. Analyse simple des caractéristiques
mean_pixel_values = np.mean(X, axis=(1, 2))
plt.figure(figsize=(10, 5))
plt.scatter(mean_pixel_values[:, 0], mean_pixel_values[:, 1], c=y, cmap='viridis')
plt.title('Distribution des classes basée sur les valeurs moyennes des pixels')
plt.xlabel('Moyenne canal Rouge')
plt.ylabel('Moyenne canal Vert')
plt.colorbar(label='Classes')
plt.show()
```

```python
#3. Analyse de l'équilibre des classes
class_counts = pd.Series(y).value_counts()
plt.figure(figsize=(10, 5))
class_counts.plot(kind='bar')
plt.title('Distribution des classes')
plt.xlabel('Classes')
plt.ylabel('Nombre d\'échantillons')
plt.show()
```
