### Prepocissing and feature

```python
#Nettoyage des données textuelles
def clean_text(text):
    if isinstance(text, str):
        # Convertir en minuscules
        text = text.lower()
        # Supprimer les caractères spéciaux
        text = re.sub(r'[^\w\s]', '', text)
        # Supprimer les chiffres
        text = re.sub(r'\d+', '', text)
    return text
data['clean_designation'] = data['designation'].apply(clean_text)
data['clean_description'] = data['description'].apply(clean_text)
```

```python
# Traitement des valeurs manquantes
# Pour les colonnes textuelles:
data['clean_description'] = data['clean_description'].fillna('pas de description')
```

```python
# Feature engineering
# Longueur des textes
data['designation_length'] = data['clean_designation'].str.len()
data['description_length'] = data['clean_description'].str.len()
```

```python
# Nombre de mots
data['designation_word_count'] = data['clean_designation'].str.split().str.len()
data['description_word_count'] = data['clean_description'].str.split().str.len()
```

```python
# Encodage des variables textuelles avec TF-IDF
tfidf_vectorizer = TfidfVectorizer(max_features=1000)
tfidf_designation = tfidf_vectorizer.fit_transform(data['clean_designation'])
tfidf_description = tfidf_vectorizer.fit_transform(data['clean_description'])
```

```python
# Convertir en DataFrame
tfidf_designation_df = pd.DataFrame(tfidf_designation.toarray(), columns=tfidf_vectorizer.get_feature_names_out())
tfidf_description_df = pd.DataFrame(tfidf_description.toarray(), columns=tfidf_vectorizer.get_feature_names_out())
```

```python
#Normalisation des features numériques
scaler = StandardScaler()
numeric_features = ['designation_length', 'description_length', 'designation_word_count', 'description_word_count']
data[numeric_features] = scaler.fit_transform(data[numeric_features])
```

```python
#Préparation du jeu de données final
X = pd.concat([data[numeric_features], tfidf_designation_df, tfidf_description_df], axis=1)
y = data['prdtypecode']

print("Shape of X:", X.shape)
print("Shape of y:", y.shape)
```

### Explications :

Nettoyage des données textuelles : Nous convertissons le texte en minuscules et supprimons les caractères spéciaux et les chiffres.
Traitement des valeurs manquantes : Nous remplaçons les descriptions manquantes par "pas de description".
Feature engineering :

Nous créons de nouvelles caractéristiques basées sur la longueur du texte et le nombre de mots.
Nous utilisons TF-IDF pour encoder les variables textuelles, ce qui nous donne une représentation numérique du texte basée sur l'importance des mots.


Normalisation : Nous normalisons les caractéristiques numériques pour qu'elles aient toutes la même échelle.
Préparation finale : Nous combinons toutes nos caractéristiques en un seul DataFrame X, et nous préparons notre variable cible y.

Après avoir exécuté ce code, vous aurez un jeu de données nettoyé et prétraité, prêt pour la modélisation.
Pourquoi ces étapes sont importantes :

Le nettoyage des données textuelles aide à réduire le bruit et à standardiser le texte.
Le traitement des valeurs manquantes évite les erreurs lors de la modélisation.
Le feature engineering nous permet de capturer plus d'informations à partir de nos données brutes.
L'encodage TF-IDF transforme le texte en format numérique que les algorithmes de ML peuvent utiliser.
La normalisation assure que toutes les caractéristiques sont à la même échelle, ce qui est important pour de nombreux algorithmes.

```python
image_path = 'C:\\Users\\TOURE\\Desktop\\Projet_Rakuten\\image_train'
df_train = pd.read_csv('X_train.csv', index_col=0)

def load_and_preprocess_image(image_id, target_size=(224, 224)):
    if image_id is None or pd.isna(image_id):
        print("Image ID is None or NaN")
        return None
    
    matching_files = [f for f in os.listdir(image_path) if f.startswith(f"image_{image_id}_")]
    
    if not matching_files:
        print(f"No matching image found for ID: {image_id}")
        return None
    
    img_path = os.path.join(image_path, matching_files[0])
    
    try:
        img = Image.open(img_path).convert('RGB')
        img = img.resize(target_size)
        img_array = img_to_array(img)
        img_array = preprocess_input(img_array)
        return img_array
    except Exception as e:
        print(f"Error processing image {image_id}: {str(e)}")
        return None

# Charger le modèle ResNet50 pré-entraîné
base_model = ResNet50(weights='imagenet', include_top=False, pooling='avg')

features = []
not_found_count = 0

# Traitement des images
for i, image_id in enumerate(df_train['imageid']):
    img_array = load_and_preprocess_image(image_id)
    if img_array is not None:
        features.append(base_model.predict(np.expand_dims(img_array, axis=0)).flatten())
    else:
        features.append(np.zeros(base_model.output_shape[1]))
        not_found_count += 1

    # Afficher la progression
    if (i + 1) % 1000 == 0:
        print(f"Processed {i + 1} images")

print(f"Nombre d'images traitées : {len(features)}")
print(f"Nombre d'images attendues : {len(df_train)}")
print(f"Nombre d'images non trouvées : {not_found_count}")

# Convertir les features en DataFrame
features_df = pd.DataFrame(features)

# Sauvegarder les features extraites
features_df.to_csv('image_features.csv', index=False)
print("Extraction des caractéristiques d'image terminée et sauvegardée.")
```

Au lieu d'utiliser les images brutes, qui contiennent beaucoup d'informations non essentielles, nous extrayons des caractéristiques de haut niveau qui capturent l'essence de l'image.
    Pourquoi utiliser un modèle pré-entraîné comme ResNet50 :
        Efficacité : ResNet50 a déjà appris à reconnaître une grande variété de caractéristiques visuelles à partir de millions d'images.
        Transfert d'apprentissage : Nous profitons de cet apprentissage pré-existant pour notre tâche spécifique.
        Gain de temps : Cela évite d'avoir à entraîner un modèle complexe à partir de zéro.
    Le processus de traitement :
        Chargement de l'image
        Redimensionnement pour correspondre à l'entrée attendue par ResNet50
        Prétraitement de l'image (normalisation)
        Passage de l'image à travers ResNet50 pour extraire les caractéristiques
